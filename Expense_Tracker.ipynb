```python

expenses = {}

expense_id = 1

def add_expense():

  global expense_id

  date = input("Enter Date: ")
  category = input("Enter Category: ")
  description = input("Enter Description: ")
  amount = float(input("Enter Amount: "))

  expenses[expense_id]={"Date": date, "Category": category, "Description": description, "Amount": amount}

  expense_id += 1

  print("Expense added successfully ")
  print("\n")


def view_expenses():

  if len(expenses)==0:
    print("No expenses found!!!\n\n")
    return

  for i,j in expenses.items():
    print("Expense_ID: ",i)
    print("Date: ", j["Date"])
    print("Category: ",j["Category"])
    print("Description: ",j["Description"])
    print("Amount: ",j["Amount"])
    print("---------------\n")


def search_expenses():
  cat = input("Enter the Category: ")

  found = False

  for i,j in expenses.items():
    if cat.lower() == j["Category"].lower():
      found = True

      print("Category: ",j["Category"])
      print("Expense_ID: ",i)
      print("Date: ", j["Date"])
      print("Description: ",j["Description"])
      print("Amount: ",j["Amount"])
      print("---------------\n")

  if not found:
      print("Category could not be found.")

def tot_expenses():
  sum = 0
  for i,j in expenses.items():
    sum = sum + int(j["Amount"])

  print("==========================================")
  print(f"Your total expense till now is Rs. {sum}")
  print("==========================================\n")

def monthly_expense():

  if len(expenses) == 0:
    print("No expenses found !!")
    return

  monthly_total ={}

  for i,j in expenses.items():
    month = j["Date"][3:]

    if month not in monthly_total:
      monthly_total[month] = j["Amount"]

    else:
      monthly_total[month] = monthly_total[month] + j["Amount"]

  for a,b in monthly_total.items():
    print(f"Total expense for month {a} is Rs.{b}")



while True:
  
  print("=======Expense Tracker System=======\n")
  print("1. Add expenses")
  print("2. View expenses")
  print("3. Search your expenses")
  print("4. Total expenses")
  print("5. Monthly expenses")
  print("6. Exit\n")

  choice = int(input("Enter your choice from(1-6): "))
  print("\n")

  if choice == 1:
    add_expense()

  elif choice == 2:
    view_expenses()

  elif choice == 3:
    search_expenses()

  elif choice == 4:
    tot_expenses()

  elif choice == 5:
    monthly_expense()
  
  elif choice == 6:
        print("Thank You")
        break
  else:
    print("Invalid choice")
```
